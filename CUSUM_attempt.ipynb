```python
import pandas as pd
import numpy as np
```

```python
df_vali = pd.read_csv('data/task_4_2.csv', sep=',')
df_vali.tail()
```

```python
df_vali.shape
```

```python
df_eval = pd.read_csv('data/task_5.csv', sep=',')
df_eval.head()
```

```python
df_eval.shape
```

```python
walmart = pd.concat([df_vali,df_eval])
walmart = walmart.loc[:,['wm_yr_wk','weekly_revenue','store_id']]
walmart.head()
```

```python
walmart.shape
```

```python
stores = [store_rev for _, store_rev in walmart.groupby('store_id')]
```

```python
stores[0].reset_index(drop=True)
```

```python
df_stores = stores[0].loc[:,['wm_yr_wk']].reset_index(drop=True)
for i in range(0,len(stores)):
    singleStore = stores[i].reset_index(drop=True)
    singleStore = singleStore.loc[:,['weekly_revenue']]
    singleStore.columns = [stores[i].iloc[0,2]]
#     print(singleStore.head())
    df_stores = pd.concat([df_stores, singleStore], axis=1)
```

```python
df_stores.head()
```

```python
df_stores.to_csv('Data/CUSUM_data.csv', index=False)
```
